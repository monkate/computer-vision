```python
%matplotlib inline 
from skimage import io, color
from matplotlib import pyplot as plt
import numpy as np
from scipy.ndimage import gaussian_filter, sobel

def conv2d(image, kernel):
    result = gaussian_filter(image, sigma=2, order=kernel)
    plt.figure(figsize=(10, 10))
    plt.subplot(1,2,1)
    plt.imshow(image, cmap='gray')
    plt.subplot(1,2,2)
    plt.imshow(result, cmap='gray')
    plt.show()
    #return result

image = color.rgb2gray(io.imread('dog.jpg'))

conv2d(image, 0) # 0 for Gaussian kernel
#plt.imshow(conv2d(image, []), cmap='gray')
```

Now, implement a simple edge detector by first filtering with a Sobel operator and then thresholding the ***RESULT***. How does smoothing the input change the ***RESULT***?

```python
result = sobel(image)
plt.figure(figsize=(10, 10))
plt.subplot(1,2,1)
plt.imshow(image, cmap='gray')
plt.subplot(1,2,2)
plt.imshow(result, cmap='gray')
plt.show()
```

```python
import cv2
ret, thresh = cv2.threshold(result, 0, 1, cv.THRESH_BINARY)
plt.imshow(thresh, 'gray')
plt.show()
```

```python
threshCopy = np.uint8(thresh)
edges = cv2.Canny(threshCopy,1,100)
plt.figure(figsize=(10, 10))
plt.subplot(121),plt.imshow(thresh,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(edges,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])

plt.show()
```
